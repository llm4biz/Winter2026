```python
import gensim.downloader as api
```

```python
wv = api.load('word2vec-google-news-300')
```

```python
# Funny/Surprising Examples
res = wv['cat'] + wv['baby']
closest_words = wv.similar_by_vector(res, topn=5)
print(closest_words)
```

```python
# Pizza is to Italy as Sushi is to...
result = wv.most_similar(positive=['sushi', 'Italy'], negative=['pizza'], topn=1)
print(result)
```

```python
wv.most_similar_cosmul(positive=['king', 'woman'], negative=['man'], topn=3)
```

```python
# Workplace bias
print(f"Young + innovative: {wv.similarity('young', 'innovative'):.3f}")
print(f"Old + innovative: {wv.similarity('old', 'innovative'):.3f}")
```
