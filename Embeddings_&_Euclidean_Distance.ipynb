```python
from google.colab import userdata
from openai import OpenAI
```

```python
client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

```python
response1 = client.embeddings.create(
    input="I am hungry",
    model="text-embedding-3-small"
)
res1 = response1.data[0].embedding

response2 = client.embeddings.create(
    input="I am looking for a snack",
    model="text-embedding-3-small"
)
res2 = response2.data[0].embedding

response3 = client.embeddings.create(
    input="I am angry",
    model="text-embedding-3-small"
)
res3 = response3.data[0].embedding

response4 = client.embeddings.create(
    input=",fsdkh vfksdjh fiusdkhf udskh fsfh kjsgdfgsjkhsdfr tgksdf gk ids fdsfg shk",
    model="text-embedding-3-small"
)
res4 = response4.data[0].embedding
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
point1 = np.array(res1).reshape(1, -1)
point2 = np.array(res2).reshape(1, -1)
point3 = np.array(res3).reshape(1, -1)
point4 = np.array(res4).reshape(1, -1)

# Calculate the Euclidean distance
# distance1_2 = np.linalg.norm(point1 - point2)
# distance1_3 = np.linalg.norm(point1 - point3)
# distance1_4 = np.linalg.norm(point1 - point4)
sim_1_2 = cosine_similarity(point1, point2)
print(sim_1_2)
sim_1_3 = cosine_similarity(point1, point3)
print(sim_1_3)
sim_1_4 = cosine_similarity(point1, point4)
print(sim_1_4)

# print(f"The Euclidean distance between point1 and point2 is: {distance1_2}")
# print(f"The Euclidean distance between point1 and point3 is: {distance1_3}")
# print(f"The Euclidean distance between point1 and point4 is: {distance1_4}")
```
