```python
from openai import OpenAI
```

```python
# Colab-specific import for secrets - OpenAI API key
from google.colab import userdata
```

```python
client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
```

# Chat Completions

```python
response = client.responses.create(
    model="gpt-5.2",
    input="Tell me a joke about humanoid robots in California."
)
```

```python
print(response.output_text)
```

```python
response2 = client.responses.create(
    model="gpt-5.2",
    previous_response_id=response.id,
    input=[{"role": "user", "content": "I did not get it! Can you explain the joke?"}],
)
```

```python
print(response2.output_text)
```

# Image Analysis

```python
response = client.responses.create(
    model="gpt-5.2",
    input=[
        {"role": "user", "content": "What is this image about?"},
        {
            "role": "user",
            "content": [
                {
                    "type": "input_image",
                    "image_url": "https://i.imgur.com/uw6RD8M.png"
                }
            ]
        }
    ]
)
```

```python
print(response.output_text)
```

```python
response.id
```

```python
follow_up = client.responses.create(
    model="gpt-4.1",
    tools=[{"type": "web_search"}],
    previous_response_id=response.id,
    input=[{"role": "user", "content": "Can you tell me who are these two instructors?"}]
)
```

```python
follow_up.output_text
```
